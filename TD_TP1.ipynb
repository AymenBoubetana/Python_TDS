```python
# Exercice 1

def isAlternate(l:list):
    isTrue = False
    for i in range(len(l)):
        if i%2 ==0:
           if i==l[i]:
               isTrue = True
           else:
               return False
        else:
           if i==-l[i]:
               isTrue =True
           else:
               return False

    return isTrue

print(isAlternate([0,-1,2,-3,4,-5]))
```

```python
# Exercice 2
def HeronIte(reel:float):
    U = reel

    for i in range(1000):
        U_new = 1/2*(U + reel/U)

        if (abs(U-U_new) <0.0001):
            return U_new
        else:
            U = U_new


nombre = 25
resultat = HeronIte(nombre)
print(f"La racine carrée de {nombre} est approximativement de la methode iterative {resultat}")


def HeronRecur(reel: float, U_old=None, U_new=None):

   


    U_new = 1 / 2 * (U_old + reel / U_old)

    # Vérifier la convergence
    if abs(U_new - U_old) < 0.0001:  # Vérifier la différence entre les estimations
        return U_new
    else:

        return HeronRecur(reel, U_new, U_new)


nombre = 25
resultat = HeronRecur(nombre)
print(f"La racine carrée de {nombre} est approximativement de la methode recurssive {resultat}")
```

```python
# Exercice 3

def SomDiv(n):
    """Retourne la somme des diviseurs propres de n (excluant n lui-même)."""
    if n < 1:
        return 0
    somme = 0
    for i in range(1, n // 2 + 1):
        if n % i == 0:  # Si i est un diviseur de n
            somme += i
    return somme


def estParfait(n):
    if SomDiv(n) == n:
        return True
    else:
        return False


def estPremier(n):
    if n<=1:
        return False

    for i in range(2,n//2 +1):
        if n % i == 0:
            return False
    return True


def estChanceux(n):
    for i in range(0,n):
        if estPremier(n+i+i**2):
            return True
    return False



L_parfaits = list()
L_chanceux = []

for i in range(2,1001):
    if estParfait(i):
        L_parfaits.append(i)
    if estChanceux(i):
        L_chanceux.append(i)

print(L_chanceux)
```

```python
# Exercice 4:

# question 1
def jumeau(N):
    """Retourne la première paire de nombres premiers jumeaux à partir de N+1."""
    for i in range(N+1, N + 1000):  
        if estPremier(i):
            p = i
            if estPremier(i + 2):  
                q = i + 2
                return [p, q]  

    return None  


print(jumeau(20))  

# question 2
def LesJumeaux(N):
    Tot = list()
    for i in range(N-1):
        if estPremier(i):
            p = i
            if estPremier(i + 2):  
                q = i + 2
                Tot.append((p, q)) 
    return Tot        
                
print(LesJumeaux(18))                           
```

```python
# Exercice 5 :
import random

# Fonction pour un tour de jeu
def jouer_tour(nom_joueur, score_actuel):
    print(f"\nTour de {nom_joueur}:")
    tour_points = 0
    while True:
        lancer = random.randint(1, 6)
        print(f"{nom_joueur} lance le dé et obtient {lancer}")

        # Si le joueur obtient un 1, il perd la main
        if lancer == 1:
            print(f"{nom_joueur} perd la main en obtenant un 1.")
            break
        # Si le joueur obtient un nombre pair, ajouter le nombre au score
        elif lancer in [2, 4, 6]:
            tour_points += lancer
            print(f"Nombre pair obtenu, {nom_joueur} gagne {lancer} points.")
        # Si le joueur obtient un 3, doubler les points
        elif lancer == 3:
            score_actuel *= 2
            print(f"Le score de {nom_joueur} est doublé à {score_actuel}!")
        # Si le joueur obtient un 5, perdre 2 points
        elif lancer == 5:
            score_actuel -= 2
            print(f"{nom_joueur} perd 2 points. Score actuel: {score_actuel}.")

        # Afficher les points de ce tour
        print(f"Points du tour: {tour_points}")
    
    score_actuel += tour_points
    return score_actuel

# Fonction principale pour le jeu
def jeu_de_des():
    score_joueur_1 = 0
    score_joueur_2 = 0
    score_cible = 50

    while score_joueur_1 < score_cible and score_joueur_2 < score_cible:
        # Tour du joueur 1
        score_joueur_1 = jouer_tour("Joueur 1", score_joueur_1)
        print(f"Score du Joueur 1: {score_joueur_1}")
        if score_joueur_1 >= score_cible:
            print("Joueur 1 gagne la partie!")
            break
        
        # Tour du joueur 2
        score_joueur_2 = jouer_tour("Joueur 2", score_joueur_2)
        print(f"Score du Joueur 2: {score_joueur_2}")
        if score_joueur_2 >= score_cible:
            print("Joueur 2 gagne la partie!")
            break

# Lancer le jeu
jeu_de_des()
```

```python
# Exercice 6:
import math

def Zeller(j, m, a, s):
    if m == 1 or m == 2:
        m += 12
        a -= 1
        
    return (j + math.floor((26 * (m + 1)) / 10) + a + math.floor(a / 4) + math.floor(s / 4) + 5 * s) % 7


def estBissextile(annee):
    return (annee % 4 == 0 and annee % 100 != 0) or (annee % 400 == 0)

# Fonction pour retourner le nombre de jours dans un mois donné
def nbJour(mois, annee):
    # Liste du nombre de jours pour chaque mois
    jours_par_mois = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    # Si c'est une année bissextile, février a 29 jours
    if mois == 2 and estBissextile(annee):
        return 29
    else:
        return jours_par_mois[mois - 1]



def afficherJourSemaine(strDate):
    # Séparation de la chaîne 'dd-mm-yyyy' en jour, mois et année
    jour, mois, annee = map(int, strDate.split('-'))

    # Tester si la date est valide
    if mois < 1 or mois > 12:
        print("Mois invalide")
        return
    if jour < 1 or jour > nbJour(mois, annee):
        print("Jour invalide")
        return

    # Calculer le jour de la semaine avec la formule de Zeller
    siecle = annee // 100
    annee_mod = annee % 100
    jour_semaine = Zeller(jour, mois, annee_mod, siecle)

    # Dictionnaire pour convertir le résultat de Zeller en jours de la semaine
    jours = ["Samedi", "Dimanche", "Lundi", "Mardi", "Mercredi", "Jeudi", "Vendredi"]

    # Afficher le jour de la semaine correspondant
    print(f"La date {strDate} correspond à un {jours[jour_semaine]}.")

# Exemple d'utilisation
strDate = input("Saisir une date (dd-mm-yyyy) : ")
afficherJourSemaine(strDate)
```

```python
# Exercice 7:

def afficher_etat_jeu(tas):
    print("\nTas       : (1, 2, 3, 4, 5, 6)")
    print(f"Allumettes: {tas}")

def choix_valide(tas, choix_tas, choix_allumettes):
    # Vérifier si le tas est valide
    if choix_tas < 1 or choix_tas > 6:
        print("Erreur : Numéro de tas invalide. Choisissez un tas entre 1 et 6.")
        return False
    # Vérifier si le tas n'est pas vide
    if tas[choix_tas - 1] == 0:
        print(f"Erreur : Le tas {choix_tas} est vide.")
        return False
    # Vérifier si le nombre d'allumettes à retirer est valide
    if choix_allumettes < 1 or choix_allumettes > tas[choix_tas - 1]:
        print(f"Erreur : Vous ne pouvez retirer que entre 1 et {tas[choix_tas - 1]} allumettes du tas {choix_tas}.")
        return False
    return True

def retirer_allumettes(tas, choix_tas, choix_allumettes):
    tas[choix_tas - 1] -= choix_allumettes

def jeu_termine(tas):
    return sum(tas) == 0

def jouer():
    # Initialiser les tas avec les allumettes
    tas = [1, 2, 3, 4, 5, 6]
    
    # Saisir les noms des joueurs
    joueur1 = input("Nom du joueur 1 : ")
    joueur2 = input("Nom du joueur 2 : ")
    
    joueurs = [joueur1, joueur2]
    tour = 0  # Variable pour alterner entre les joueurs
    
    while not jeu_termine(tas):
        # Afficher l'état actuel du jeu
        afficher_etat_jeu(tas)
        
        # Déterminer le joueur en cours
        joueur_actuel = joueurs[tour % 2]
        print(f"\n-------- Prochain joueur : {joueur_actuel} --------")
        
        # Demander au joueur de choisir un tas et un nombre d'allumettes
        choix_tas = int(input(f"{joueur_actuel}, choisissez un tas (1-6) : "))
        choix_allumettes = int(input(f"Combien d'allumettes voulez-vous retirer du tas {choix_tas} ? "))
        
        # Vérifier si le choix est valide
        while not choix_valide(tas, choix_tas, choix_allumettes):
            choix_tas = int(input(f"{joueur_actuel}, choisissez un tas (1-6) : "))
            choix_allumettes = int(input(f"Combien d'allumettes voulez-vous retirer du tas {choix_tas} ? "))
        
        # Retirer les allumettes du tas choisi
        retirer_allumettes(tas, choix_tas, choix_allumettes)
        
        # Vérifier si le jeu est terminé
        if jeu_termine(tas):
            afficher_etat_jeu(tas)
            print(f"\nLe joueur {joueur_actuel} a pris la dernière allumette et a perdu !")
            autre_joueur = joueurs[(tour + 1) % 2]
            print(f"Le joueur {autre_joueur} est le gagnant.")
            break
        
        # Passer au tour du joueur suivant
        tour += 1

# Lancer le jeu
jouer()
```
