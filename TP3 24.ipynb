```python
# Exercice 1:

class Robot:
    def __init__(self,nom,x=0,y=0,direction="Est"):
        assert isinstance(nom,str) and len(nom)!=0
        assert isinstance(direction,str) and direction in ["Nord", "Est", "Sud","Ouest"]
        self.nom = nom
        self.__position = {"x":x,"y":y}
        self.__direction = direction
    
    # Getter pour accéder à l'attribut privé __position
    def get_position(self):
        return self.__position
    
    # Setter pour définir l'attribut privé __position
    def set_position(self, position): 
        self.__position = position
    
    # Getter pour accéder à l'attribut privé __direction
    def get_direction(self):
        return self.__direction
    
    # Setter pour définir l'attribut privé __direction
    def set_direction(self, direction): 
        self.__direction = direction
    
    position = property(get_position, set_position)
    direction = property(get_direction, set_direction)

    def __str__(self):
        return "Nom: {0} position: ({1},{2}) Direction {3}".format(self.nom,self.__position["x"],self.__position["y"],self.__direction)
    
    def __add__(self, pas):
        """Avancer selon la direction du pas."""
        if self.direction == "Nord":
            self.__position['y'] += pas
        elif self.direction == "Sud":
            self.__position['y'] -= pas
        elif self.direction == "Est":
            self.__position['x'] += pas
        elif self.direction == "Ouest":
            self.__position['x'] -= pas
        return self

    def droite(self):
        """Tourner à droite de 90°."""
        directions = ["Nord", "Est", "Sud", "Ouest"]
        current_index = directions.index(self.direction)
        self.direction = directions[(current_index + 1) % 4]
        return self
    
    
    

r =Robot("Aymen",2,5,"Nord")
print(r.position)
```

```python
# Exercice 2:

class RobotNG(Robot):
    
    def __init__(self, nom, x=0, y=0, direction="Est", Turbo=False):
        super().__init__(nom,x=x, y=y, direction=direction)
        self.nom = nom
        self.Turbo = Turbo
    
    def __add__(self, pas):
        """Avancer selon la direction du pas."""
        if self.Turbo:
            if self.direction == "Nord":
                self._Robot__position['y'] += pas * 3  # Y augmente
            elif self.direction == "Sud":
                self._Robot__position['y'] -= pas * 3  # Y diminue
            elif self.direction == "Est":
                self._Robot__position['x'] += pas * 3  # X augmente
            elif self.direction == "Ouest":
                self._Robot__position['x'] -= pas * 3  # X diminue
        else:
            if self.direction == "Nord":
                self._Robot__position['y'] += pas  # Y augmente
            elif self.direction == "Sud":
                self._Robot__position['y'] -= pas  # Y diminue
            elif self.direction == "Est":
                self._Robot__position['x'] += pas  # X augmente
            elif self.direction == "Ouest":
                self._Robot__position['x'] -= pas  # X diminue
        return self
    
    def gauche(self):
        """Tourner à gauche de 90°."""
        directions = ["Nord", "Ouest", "Sud", "Est"]
        current_index = directions.index(self.direction)
        self.direction = directions[(current_index + 1) % 4]
        return self
    
    def __str__(self):
        return "Nom: {0} position: ({1},{2}) Direction {3} le mode est turbo: {4}".format(
            self.nom,
            self._Robot__position["x"],
            self._Robot__position["y"],
            self.direction,
            self.Turbo
        )

```

```python
def test_robot():
    print("=== Test de la classe Robot ===")
    
    # Création d'un robot
    robot = Robot("Robot")
    print(robot)  # Affiche la position initiale

    # Test de la méthode droite
    robot.droite()
    print(robot)  # Vérifie que la direction est maintenant Est

    # Test d'avancer vers l'Est
    robot + 3  # Avancer de 3 pas
    print(robot)  # Vérifie la nouvelle position

    # Test d'avancer vers le Nord
    robot + 2  # Avancer de 2 pas
    print(robot)  # Vérifie la nouvelle position

    # Test de tourner à droite à nouveau
    robot.droite()
    print(robot)  # Vérifie que la direction est maintenant Est

    # Test d'avancer vers l'Est
    robot + 1  # Avancer de 1 pas
    print(robot)  # Vérifie la nouvelle position



    # Test d'avancer vers le Nord
    robot + 5  # Avancer de 5 pas
    print(robot)  # Vérifie la nouvelle position


    print(robot)  # Vérifie que la direction est maintenant Sud
    robot + 4  # Avancer de 4 pas
    print(robot)  # Vérifie la nouvelle position

    # Test de tourner à droite et avancer
    robot.droite()  # Tourner à droite pour aller vers Ouest
    print(robot)  # Vérifie que la direction est maintenant Ouest
    robot + 3  # Avancer de 3 pas
    print(robot)  # Vérifie la nouvelle position

# Exécution du test
test_robot()
```
